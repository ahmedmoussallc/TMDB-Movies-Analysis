```python
# Importing the necessary libraries and the dataset:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv("tmdb-movies.csv")
```

```python
# Viewing the first few lines of the dataset:

df.head()
```

```python
# checking the size of the dataset:

df.shape
```

```python
# getting general info about the dataset:

df.info()
```

```python
# getting some statistical information about the dataset:

df.describe()
```

```python
# As we can see above there are some values of zeros, I'll change them to null and then drop all null values:

df.replace(0, np.nan, inplace= True)
df.describe()
```

```python
# now let's drop null values from the rest of the dataset:

df.dropna(inplace = True)
```

```python
#let's take a look at the dataset now:

df.info()
```

```python
# now let's check for duplicate values:

df.duplicated().sum()
```

```python
# now let's check if all the columns have the proper data types:

df.info()
```

```python
#The id column has a datatype of int, we should change it to string

df.id = df.id.astype(str)
type(df.id[0])
```

```python
# The release_date column is in a string format, we need to change it to date format:

df.release_date = pd.to_datetime(df.release_date)
type(df.release_date[0])
```

```python
#extracting release year from the release date column:

df.release_year = df.release_date.dt.year
```

```python
# Now after dropping null and duplicate values, adjusting data types
# Let's explore the dataset:

df.describe()
```

```python
# we can see an outlier in the release year column, as we're now in 2023
# I'll remove all rows with values > 2023

df = df[df.release_year <= 2023]
df.describe()
```

```python
# Now after the data is cleaned, it's time for us to ask some questions!
```

## 1 - What is happening to the movie rating performance over the years?

```python
#Let's explore the the movie rating overtime performance with a line plot:

df.groupby(by = "release_year")["vote_average"].mean().plot();
plt.xlabel("Movie Release Year")
plt.ylabel("Averge Vote")
plt.title("Voting Average Over The Years")
plt.show()
```

As we can see from the line chart above, the voting average seems to be going down over the years

## 2 - Is the movie business making more profit over the years or less?

```python
# In order to do this
# we need to create a new column, let's call it profit:

df["profit"] = df.revenue - df.budget
```

```python
# let's also create an adjusted profit column:

df["profit_adj"] = df.revenue_adj - df.budget_adj
```

```python
# Now let's see how the profit and the adjusted profit looking over the years:

df.groupby("release_year")["profit"].mean().plot()
plt.xlabel("Movie Release Year")
plt.ylabel("Average Profit")
plt.title("Average Profit Over The Years")
plt.show()
```

```python
df.groupby("release_year")["profit_adj"].mean().plot()
plt.xlabel("Movie Release Year")
plt.ylabel("Average Adjusted Profit")
plt.title("Average Adjusted Profit Over The Years")
plt.show()
```

In conclusion of the above two chart, it seems like the movie business is making less profit over the years

## 3 - Are the number of movies increasing per year?

```python
movies_per_year = df["release_year"].value_counts().sort_index()
movies_per_year.plot(kind = "barh", figsize = (5,10), legend = True)
plt.title("Number of movies released per year")
plt.xlabel("Number of Movies")
plt.ylabel("Release Year")
plt.show()
```

It seems like the number of movies released increase on a yearly basis, makes sense!
The Year with most movies released is 2011.
